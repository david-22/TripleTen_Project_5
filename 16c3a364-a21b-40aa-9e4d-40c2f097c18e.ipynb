## Agregar datos por usuario

[Ahora que los datos están limpios, agrega los datos por usuario y por periodo para que solo haya un registro por usuario y por periodo. Esto facilitará mucho el análisis posterior.]

```python
print(megaline_calls.head())
megaline_calls['call_date'] = pd.to_datetime(megaline_calls['call_date'])
megaline_calls['month'] = megaline_calls['call_date'].dt.month
numberof_calls = megaline_calls.groupby(['user_id', 'month']).agg({'id':'count'})
numberof_calls.reset_index(inplace=True)# Calcula el número de llamadas hechas por cada usuario al mes. Guarda el resultado.
numberof_calls = numberof_calls.rename(columns  = {'id': 'no_calls'})
print(numberof_calls)
```

```python
minutes_used = megaline_calls.groupby(['user_id', 'month']).agg({'duration': 'sum'})
minutes_used.reset_index(inplace=True)# Calcula la cantidad de minutos usados por cada usuario al mes. Guarda el resultado.
minutes_used.rename(columns={'duration': 'minutes_used'}, inplace=True)
print(minutes_used)
```

```python
print(megaline_messages.head())# Calcula el número de mensajes enviados por cada usuario al mes. Guarda el resultado. 

```

```python
print(megaline_internet.head())# Calcula el volumen del tráfico de Internet usado por cada usuario al mes. Guarda el resultado.
megaline_internet['session_date'] = pd.to_datetime(megaline_internet['session_date'])
megaline_internet['month'] = megaline_internet['session_date'].dt.month
internet_traffic = megaline_internet.groupby(['user_id', 'month']).agg({'mb_used': 'sum'})
internet_traffic.reset_index(inplace=True)
print(internet_traffic)

internet_traffic = internet_traffic.rename(columns = {'mb_used': 'gygabytes_used'})
internet_traffic['gygabytes_used'] = np.ceil(internet_traffic['gygabytes_used']/1024)
print(internet_traffic)
```

[Junta los datos agregados en un DataFrame para que haya un registro que represente lo que consumió un usuario único en un mes determinado.]

```python
megaline_merged = numberof_calls.merge(minutes_used, on=['user_id', 'month'], how='outer').merge(sent_messages, on=['user_id', 'month'], how='outer').merge(internet_traffic, on=['user_id', 'month'], how='outer')# Fusiona los datos de llamadas, minutos, mensajes e Internet con base en user_id y month
megaline_merged.fillna(0, inplace=True)
print(megaline_merged)
megaline_merged['minutes_used'] = np.ceil(megaline_merged['minutes_used'])
print(megaline_merged)
```

```python
# Añade la información de la tarifa

megaline_users.rename(columns={'plan': 'plan_name'}, inplace=True)

new_merged = megaline_merged.merge(megaline_users, on='user_id', how='outer').merge(megaline_plans, on='plan_name', how='outer')

new_merged.dropna(subset=['month'],inplace=True)

new_merged['churn_date'].fillna('current', inplace=True)
new_merged['mb_per_month_included'] = new_merged['mb_per_month_included']/1024
print(new_merged)
```

[Calcula los ingresos mensuales por usuario (resta el límite del paquete gratuito del número total de llamadas, mensajes de texto y datos; multiplica el resultado por el valor del plan de llamadas; añade la tarifa mensual en función del plan de llamadas). Nota: Dadas las condiciones del plan, ¡esto podría no ser tan trivial como un par de líneas! Así que no pasa nada si dedicas algo de tiempo a ello.]

```python
def minutes_cost(row):
    if (row['plan_name'] == 'ultimate' and row['minutes_used']<= row['minutes_included']):
        return 0
    elif (row['plan_name'] == 'surf' and row['minutes_used'] <= row['minutes_included']):
        return 0
    else:
        return (row['minutes_used']-row['minutes_included'])*row['usd_per_minute']# Calcula el ingreso mensual para cada usuario

def messages_cost(row):
    if(row['plan_name'] == 'ultimate' and row['messages_sent'] <= row['messages_included']):
        return 0
    elif (row['plan_name'] == 'surf' and row['messages_sent'] <= row['messages_included']):
        return 0
    else:
        return (row['messages_sent']-row['messages_included'])*row['usd_per_message']

def megabytes_cost(row):
    if(row['plan_name'] == 'ultimate' and row['gygabytes_used'] <= row['mb_per_month_included']):
        return 0
    elif (row['plan_name'] == 'surf' and row['gygabytes_used'] <= row['mb_per_month_included']):
        return 0
    else:
        return (row['gygabytes_used']-row['mb_per_month_included'])*row['usd_per_gb']

def total_cost(row):
    return row['usd_monthly_pay'] + row['total_cost_minutes'] + row['total_cost_messages'] + row['total_cost_gygabytes']
    
new_merged['total_cost_minutes'] = new_merged.apply(minutes_cost, axis=1)
new_merged['total_cost_messages'] = new_merged.apply(messages_cost, axis=1)
new_merged['total_cost_gygabytes'] = new_merged.apply(megabytes_cost, axis=1)
new_merged['total'] = new_merged.apply(total_cost, axis=1)

print(new_merged)
```

## Prueba las hipótesis estadísticas

[Prueba la hipótesis de que son diferentes los ingresos promedio procedentes de los usuarios de los planes de llamada Ultimate y Surf.]

[Elabora las hipótesis nula y alternativa, escoge la prueba estadística, determina el valor alfa.]

```python
# Prueba las hipótesis
stats = new_merged.loc[:, ['total', 'plan_name', 'month']]
ultimate_stats = stats[stats['plan_name']=='ultimate']
surf_stats = stats[stats['plan_name'] == 'surf']

alpha = 0.05
estadistico, p_valor = st.levene(ultimate_stats['total'], surf_stats['total'])

if (p_valor < alpha):
    print('Parametro equal_var de la prueba set to True')
else:
    print('Parametro equal_var de la prueba set to False')


results = st.ttest_ind(ultimate_stats['total'], surf_stats['total'])
print('valor p: ', results.pvalue)

if(results.pvalue < alpha):
    print('Rechazamos la hipotesis nula')
else:
    print('No podemos rechazar la hipotesis nula')
```
